**HANDLING OUTLIERS USING LOG TRANSFORMATION**

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

data_cleaned['Sales_log'] = np.log(data_cleaned['Sales'] + 1)


fig = plt.figure(figsize=(12, 4))
ax1 = fig.add_subplot(121)
stats.probplot(data_cleaned['Sales'], dist="norm", plot=ax1)
ax1.set_title('Outliers before Log transformation')

ax2 = fig.add_subplot(122)
stats.probplot(data_cleaned['Sales_log'], dist="norm", plot=ax2)
ax2.set_title('Outliers after Log transformation')
plt.show()
pd.set_option('display.max_columns', None)
data_cleaned.sample(2)

```

```python
data_cleaned['Ship Mode_encoded'] = data_cleaned['Ship Mode'].replace({
    'Standard Class': 1,
    'First Class': 2,
    'Second Class': 3,
    'Same Day': 4
})
data_cleaned.sample()
```



**Train Test split**

```python
from sklearn.preprocessing import LabelEncoder

X = data_cleaned.drop(['Order Date', 'Ship Date', 'Sales', 'Sales_log'], axis=1)

# Apply LabelEncoder to categorical columns
for col in X.columns:
    if X[col].dtype == 'object':  # Check if the column is of type object (categorical)
        label = LabelEncoder()
        X[col] = label.fit_transform(X[col].values)

# Define the target variable, which is 'Sales_log'
y = data_cleaned['Sales_log']
```

**FEATURE EXTRACTION USING PCA AND STANDING SCALER**

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Scale data before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

# Now split the data
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
def train_and_evaluate(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{model.__class__.__name__} - MSE: {mse:.2f}, R2: {r2:.2f}")
```

```python

models = [
    LinearRegression(),
    Ridge(alpha=1.0),
    Lasso(alpha=0.1),
    DecisionTreeRegressor(max_depth=5),
    RandomForestRegressor(n_estimators=100, random_state=42),
    SVR(kernel='rbf')
]

for model in models:
    train_and_evaluate(model, X_train, y_train, X_test, y_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
def evaluate_regression_model(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f"R^2: {r2:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score


models = [
    LinearRegression(),
    Ridge(alpha=1.0),
    Lasso(alpha=0.1),
    DecisionTreeRegressor(max_depth=5),
    RandomForestRegressor(n_estimators=100, random_state=42),
    SVR(kernel='rbf')
]


model_names = []
r2_scores = []
mse_scores = []
rmse_scores = []


for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)


    model_names.append(model.__class__.__name__)
    r2_scores.append(r2_score(y_test, y_pred))
    mse_scores.append(mean_squared_error(y_test, y_pred))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))

    print(f"Evaluating: {model.__class__.__name__}")
    print(f"R^2: {r2_scores[-1]:.4f}, MSE: {mse_scores[-1]:.4f}, RMSE: {rmse_scores[-1]:.4f}")
    print("-" * 30)  # Separator for readability

fig, ax = plt.subplots(3, 1, figsize=(10, 15))


ax[0].barh(model_names, r2_scores, color='skyblue')
ax[0].set_title('R Scores by Model')
ax[0].set_xlabel('R Score')

ax[1].barh(model_names, mse_scores, color='lightgreen')
ax[1].set_title('MSE Scores by Model')
ax[1].set_xlabel('MSE Score')


ax[2].barh(model_names, rmse_scores, color='salmon')
ax[2].set_title('RMSE Scores by Model')
ax[2].set_xlabel('RMSE Score')

plt.tight_layout()
plt.show()
```

**HYPERPARAMETER TUNING THE BEST MODEL**

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}


grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, scoring='neg_mean_squared_error', verbose=2)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score (MSE):", -grid_search.best_score_)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

optimal_params = {'max_depth': 20, 'min_samples_leaf': 4, 'min_samples_split': 2, 'n_estimators': 300}


optimal_rf = RandomForestRegressor(
    max_depth=optimal_params['max_depth'],
    min_samples_leaf=optimal_params['min_samples_leaf'],
    min_samples_split=optimal_params['min_samples_split'],
    n_estimators=optimal_params['n_estimators'],
    random_state=42
)

optimal_rf.fit(X_train, y_train)


y_pred = optimal_rf.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r_squared = r2_score(y_test, y_pred)

print(f"Evaluated Metrics for RandomForestRegressor after Hyperparameter Tuning:")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R-squared: {r_squared:.4f}")
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import Ridge
from sklearn.svm import SVR

# Define the base models and the meta model
estimators = [
    ('rf', RandomForestRegressor(n_estimators=10, random_state=42)),
    ('svr', SVR(kernel='linear'))
]
stack_model = StackingRegressor(estimators=estimators, final_estimator=Ridge())

# Train the stacking model
stack_model.fit(X_train, y_train)

# Predict and evaluate
y_pred_stack = stack_model.predict(X_test)
evaluate_regression_model(y_test, y_pred_stack)
```
